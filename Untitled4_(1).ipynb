## **Does price vary with Airlines?**

```python
plt.figure(figsize=(15,5))
sns.boxplot(x=df["airline"],y=df["price"])
plt.xlabel('Airline',fontsize=15)
plt.ylabel('Price',fontsize=15)
plt.show()
```

## **How is the price affected when tickets are bought in just 1 or 2 days before departure?**

```python
plt.figure(figsize = (16, 8))
sns.lineplot(data = df, x = 'days_left', y= 'price')
plt.title('Ticket Price Versus Days Left')
plt.show()
```

```python
plt.figure(figsize = (16, 8))
sns.lineplot(data = df, x = 'days_left', y= 'price', hue = 'airline',palette="nipy_spectral")
plt.title('Ticket Price Versus Days Left')
plt.show()
```

in the 2 previous plot you see the inverse between days left and price in airlines.



**Conclusion :**
when you increase the days left the departure , the price may decrease.

## **Does ticket price change based on the departure time and arrival time?**

```python
plt.figure(figsize = (15, 7))
plt.subplot(1,2,1)
sns.boxplot(x=df["departure_time"], y=df["price"],palette="Greens_d")
plt.subplot(1,2,2)
sns.boxplot(x=df["arrival_time"], y=df["price"],palette="Blues_d")
plt.show()
```

I think yes , look at price in late night and night in poth : departure and arrival.

## **How the price changes with change in Source and Destination?**

```python
plt.figure(figsize = (15, 7))
plt.subplot(1,2,1)
sns.kdeplot(hue='source_city',x="price",data=df,palette="seismic")
plt.subplot(1,2,2)
sns.kdeplot(hue='destination_city',x="price",data=df,palette="seismic")
```

this change by passengers , in our data show a lot of pessengers make flight from Delhi and to Mumbai but price based on the pessengers needs

## **How does the ticket price vary between Economy and Business class?**

```python
pie_class=[ec,bes]
names=["Economy","Business"]
colors=["#4766A1","#EA906C"]
plt.pie(pie_class,labels=names,colors=colors,wedgeprops = { 'linewidth' : 3, 'edgecolor' : 'white' })
plt.show()
```
